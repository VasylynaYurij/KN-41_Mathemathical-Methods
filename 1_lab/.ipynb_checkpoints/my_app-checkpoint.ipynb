```python
import matplotlib.pyplot as plt  # Імпортуємо бібліотеку matplotlib
import numpy as np  # Імпортуємо numpy, якщо потрібен для роботи

# Створюємо графік
plt.plot([1, 2, 4, 8])
plt.show()  # Показуємо графік
```

```python
s = 100  # Кількість точок
rng = np.random.default_rng(seed=101)  # Генератор випадкових чисел

# Генерація випадкових значень для x та y
x = rng.integers(0, 5, size=s, endpoint=True)
y = rng.integers(0, 100, size=s, endpoint=True)

# Виведення статистики
print(f"""
X має {x.size} елементів, з них унікальні {np.unique(x)}
По Х: Максимум {x.max()}, Мінімум {x.min()}, Середнє значення: {x.mean()}

Y має {y.size} елементів, з них унікальних {np.unique(y).size} (не будемо виводити всі елементи)
По Y: Максимум {y.max()}, Мінімум {y.min()}, Середнє значення: {y.mean()}
""")

# Побудова графіка
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Випадкові величини х та у')

# Додавання горизонтальної лінії, що показує середнє значення y
plt.plot(np.repeat(y.mean(), x.max()+1), "--")

plt.show()
```

```python
# Виведення гістограми
print(np.histogram(y))
plt.hist(y)
plt.show()
```

```python
x = np.linspace(0, 5, 100)
ya = (5 - x) / 4
yb = (3 - x) ** 2
yc = (2 - x)
yd = np.full_like(x, 1)

plt.plot(x, ya, label='y = (5 - x)/4')
plt.plot(x, yb, label='y = (3 - x)**2')
plt.plot(x, yc, label='y = 2 - x')
plt.plot(x, yd, label='y = 1')

plt.title('Візуалізація')
plt.xlabel('x')
plt.ylabel('y')

# Додаємо легенду
plt.legend()

plt.show()
```
